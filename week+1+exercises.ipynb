```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(1001)
x
```

```python
y = np.log10(x+1)
y
```

```python
plt.plot(x,y,label = "testing")
plt.show()
```

```python
plt.plot(x,y)
plt.title('Graph of y = log(x+1)')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.grid()
plt.show()
```
